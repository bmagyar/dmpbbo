```python
%pylab
%matplotlib inline
```

```python
dmp = dmpbbo.Dmp(2, 3)
```

```python
w = [-40., 20., -20., 0., 0., 0.]
duration = 1.
n_steps = 1000
dmp.set_tau(duration)
dmp.set_initial_state([0., 0.])
dmp.set_attractor_state([1., 1.])
traj = dmp.trajectory(n_steps, w)
```

```python
for i, x in enumerate(traj):
    subplot (1, 4, i + 1)
    plot(traj[0], x)
    #axis('normal')
```

```python
dmp.set_initial_state([1., -1.])
dmp.set_attractor_state([-1., 2.])
duration = 5.
dmp.set_tau(duration)
```

```python
traj = dmp.trajectory(n_steps, w)
ts =linspace(0., duration, n_steps)
for i, x in enumerate(traj):
    subplot (1, 4, i + 1)
    plot(traj[0], x)
    axis('auto')
```
